# Humming plot

```python
import math

def somme_choix(n: int, w: int) -> int:
    """Calcule Σ_{k=0..w} C(n, k)."""
    w = min(w, n)  # on borne w à n
    return sum(math.comb(n, k) for k in range(w + 1))
```

```python
# For a max number of elements in n=5 and w=2
print(somme_choix(n=100, w=2))
```

```python
import numpy as np

import matplotlib.pyplot as plt
import sys
sys.path.append('../src')
import utils

# Plot the humming values for n=100 and w from 1 to 20

n = list(range(1, 101))
w_values = list(range(1, 10))
humming_values = {w: [somme_choix(ni, w) for ni in n] for w in w_values}
for w, values in humming_values.items():
    plt.plot(n, values, label=f'w={w}')
plt.yscale('log')
plt.xlabel('n (number of tokens)')
plt.ylabel('Humming value Σ_{k=0..w} C(n, k)')
plt.title('Humming values for different w')
plt.legend()
plt.grid(True)
plt.show()
```
