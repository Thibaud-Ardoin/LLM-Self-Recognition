## Comparaisons

```python
# Calculate average quality for vanilla
df_vanilla['Avg_Quality'] = df_vanilla['Quality'].apply(lambda x: np.mean(x))
df_vanilla['Avg_ppl'] = df_vanilla['Perplexity'].apply(lambda x: np.mean(x))

# Calculate average quality for steered
df_steered['Avg_Quality'] = df_steered['Quality'].apply(lambda x: np.mean(x))
df_steered['Avg_ppl'] = df_steered['Perplexity'].apply(lambda x: np.mean(x))

# Merge dataframes on Model and Dataset
comparison = df_steered.merge(
    df_vanilla[['Model', 'Dataset', 'Avg_Quality', 'Avg_ppl']], 
    on=['Model', 'Dataset'], 
    suffixes=('_steered', '_vanilla')
)

# Calculate difference
comparison['Quality_Difference'] = comparison['Avg_Quality_steered'] - comparison['Avg_Quality_vanilla']
comparison['PPL_Difference'] = comparison['Avg_ppl_steered'] - comparison['Avg_ppl_vanilla']

comparison['Quality_Difference_percentage'] = (comparison['Quality_Difference'] / comparison['Avg_Quality_vanilla']) * 100
comparison['PPL_Difference_percentage'] = (comparison['PPL_Difference'] / comparison['Avg_ppl_vanilla']) * 100


comparison = comparison[comparison["noise_max"]==5]
# Display comparison
comparison[['Model', 'Dataset', 'noise_type', 'noise_max', 'Avg_Quality_vanilla', 'Avg_Quality_steered', 'Quality_Difference_percentage', "Avg_ppl_vanilla", "Avg_ppl_steered", 'PPL_Difference_percentage', "mean_text_accuracy"]]
```
